```
# ── BONUS : Tester en direct dans Colab ──────────────────────
from transformers import pipeline

khal = pipeline('text-generation', model='khal_modele_entraine')

def demander_khal(question):
    prompt = f'<|prompt|>{question}<|completion|>'
    resultat = khal(prompt, max_new_tokens=100, do_sample=True,
                    temperature=0.7, pad_token_id=khal.tokenizer.eos_token_id)[0]['generated_text']
    if '<|completion|>' in resultat:
        reponse = resultat.split('<|completion|>')[1].split('<|end|>')[0].strip()
    else:
        reponse = resultat[len(prompt):]
    return reponse

# Test
print(demander_khal('Bonjour !'))
print(demander_khal('Qui es-tu ?'))
print(demander_khal("C'est quoi l'IA ?"))
```
